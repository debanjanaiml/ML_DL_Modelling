## Anomaly Detection using AutoEncoders

* An autoencoder is a special type of neural network that is trained to copy its input to its output. 
* For example, given an image of a handwritten digit, an autoencoder first encodes the image into a lower dimensional latent representation, then decodes the latent representation back to an image. 
* An autoencoder learns to compress the data while minimizing the reconstruction error.

##### Import the required libraries

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, losses
from tensorflow.keras.models import Model

import matplotlib.pyplot as plt
%matplotlib inline
```

##### TF Session Initialization

```python
print("TensorFlow Version: ", tf.__version__)
tf.enable_eager_execution()
# For lower Tensorflow version
# init = (tf.global_variables_initializer(), tf.local_variables_initializer())
# sess = tf.Session()
# sess.run(init)
```

##### Import the dataset

Here we use the ECG5000 dataset which contains 5000 ECG images 

```python
df = pd.read_csv('http://storage.googleapis.com/download.tensorflow.org/data/ecg.csv', header=None)
raw_data = df.values
df.head()
```

```python
df.describe()
```

```python
# The last element contains the labels
labels = raw_data[:, -1]

# The other data points are the electrocadriogram data
data = raw_data[:, 0:-1]

train_data, test_data, train_labels, test_labels = train_test_split(
                        data, labels, test_size=0.2, random_state=365)
```

##### Normalize the data between 0 and 1

```python
min_val = tf.reduce_min(train_data)
max_val = tf.reduce_max(train_data)

train_data = (train_data - min_val) / (max_val - min_val)
test_data = (test_data - min_val) / (max_val - min_val)

train_data = tf.cast(train_data, tf.float32)
test_data = tf.cast(test_data, tf.float32)
```

```python
print("Training Data After Normalization: \n", train_data)
```

##### Seperate out the Normal Data from the Anomolous Data

```python
# Create binary index from labels for filtering 
train_labels = train_labels.astype(bool)
test_labels = test_labels.astype(bool)
```

#### Normal ECG

```python
# Filter Normal ECG data based on binary indices from labels
# if labels = 1, it means the data was from a normal ECG
normal_train_data = train_data[train_labels]
normal_test_data = test_data[test_labels]

# print Normal ECG
print("Normal ECG Data: \n", normal_train_data)
```

Here we can see that 2338 images of the 3999 got filtered as normal ECG

##### Plot any Normal ECG image

```python
plt.grid()
plt.plot(np.arange(140), normal_train_data[0].numpy())
plt.title("A Normal ECG")
plt.show()
```

#### Anomalous ECG

```python
anomalous_train_data = train_data[~train_labels]
anomalous_test_data = test_data[~test_labels]

# print Anomalous ECG
print("Anomalous ECG Data: \n", anomalous_train_data)
```

##### Plot any Anomalous ECG image

```python
plt.grid()
plt.plot(np.arange(140), anomalous_train_data[0].numpy())
plt.title("An Anomalous ECG")
plt.show()
```

### Model

```python
class AnomalyDetector(Model):
    def __init__(self):
        super(AnomalyDetector, self).__init__()
        self.encoder = tf.keras.Sequential([
          layers.Dense(32, activation="relu"),
          layers.Dense(16, activation="relu"),
          layers.Dense(8, activation="relu")])

        self.decoder = tf.keras.Sequential([
          layers.Dense(16, activation="relu"),
          layers.Dense(32, activation="relu"),
          layers.Dense(140, activation="sigmoid")])

    def call(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

autoencoder = AnomalyDetector()
```

##### Model Parameters

```python
autoencoder.compile(optimizer='adam', loss='categorical_crossentropy')
```

##### Train Model

```python
history = autoencoder.fit(normal_train_data, 
                          normal_train_data, 
                          epochs=1024, 
                          batch_size=512, 
                          validation_data=(test_data, test_data),
                          shuffle=True
                         )
```

##### Plot Loss Curve

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
```

### Encoder - Decoder

```python
encoded_imgs = autoencoder.encoder(normal_test_data).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()
```

##### Plot resultant dataframe from encoding - decoding

```python
plt.plot(normal_test_data[0].numpy(), 'b')
plt.plot(decoded_imgs[0], 'r')
plt.fill_between(np.arange(140), decoded_imgs[0], normal_test_data[0].numpy(), color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

##### Plotting Encoded-Decoded variance on Anomalous Data

```python
encoded_imgs = autoencoder.encoder(anomalous_test_data).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()

plt.plot(anomalous_test_data[0].numpy(), 'b')
plt.plot(decoded_imgs[0], 'r')
plt.fill_between(np.arange(140), decoded_imgs[0], anomalous_test_data[0].numpy(), color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

### Detect Anomalies


    Detect anomalies by calculating whether the reconstruction loss is greater than a fixed threshold.
    Here, we calculate the mean absolute error for normal examples from the training set, then classify future examples as anomalous if the reconstruction error is higher than one standard deviation from the training set

```python
reconstructions = autoencoder.predict(normal_train_data)
train_loss = tf.keras.losses.mae(reconstructions, normal_train_data)

plt.hist(train_loss[None,:], bins=50)
plt.xlabel("Train loss")
plt.ylabel("No of examples")
plt.show()
```

#### Threshold

```python
threshold = np.mean(train_loss) + 2*np.std(train_loss)
print("Threshold: ", threshold)
```

#### Reconstruction

```python
reconstructions = autoencoder.predict(anomalous_test_data)
test_loss = tf.keras.losses.mae(reconstructions, anomalous_test_data)

plt.hist(test_loss[None, :], bins=50)
plt.xlabel("Test loss")
plt.ylabel("No of examples")
plt.show()
```

### Classification

Classify ECG as normal or anomalous

```python
def predict(model, data, threshold):
    reconstructions = model(data)
    loss = tf.keras.losses.mae(reconstructions, data)
    return tf.math.less(loss, threshold)

def print_stats(predictions, labels):
    print("Accuracy = {}".format(accuracy_score(labels, preds)))
    print("Precision = {}".format(precision_score(labels, preds)))
    print("Recall = {}".format(recall_score(labels, preds)))
```

### Prediction and Metrics

```python
preds = predict(autoencoder, test_data, threshold)
print_stats(preds, test_labels)
```
