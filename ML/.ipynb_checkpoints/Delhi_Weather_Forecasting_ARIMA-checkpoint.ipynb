## Delhi Weather Forecasting (ARIMA)

This dataset contains the temperature and weather forecasting details for Delhi. 


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
weather = pd.read_csv('./data/delhi_weather.csv', parse_dates=['datetime_utc'], index_col='datetime_utc')
weather.head()
```

```python
weather.info()
```

So there are a lot of missing values and many features which either have missing/NaN values and other data which we do not know.
A wise choice here to select only particular features we are confident that might be more useful, and drop other features

```python
weather1 = weather.iloc[:,:13]
weather1.head()
```

```python
# Checking for NaN values
weather1.isna().sum()
```

```python
# lets rename the features
weather1.columns
weather1 = weather1.rename(index=str, columns={' _conds': 'condition', 
                                               ' _dewptm': 'dew', 
                                               ' _fog': 'fog', 
                                               ' _hail': 'hail', 
                                               ' _heatindexm': 'heat_index', 
                                               ' _hum': 'humidity',
                                               ' _precipm': 'precipitation', 
                                               ' _pressurem': 'pressure', 
                                               ' _rain': 'rain', 
                                               ' _snow': 'snow', 
                                               ' _tempm': 'temperature', 
                                               ' _thunder': 'thunder',
                                               ' _tornado': 'tornado'
                                              })
weather1.head()
```

```python
# lets drop the columns heat_index and precipitation
drop_cols = ['heat_index', 'precipitation']
weather1.drop(labels=drop_cols, axis=1, inplace=True)
weather1.info()
```

```python
# Checking for NaN values
weather1.isna().sum()
```

```python
# drop all the records with missing values
weather1.dropna(inplace=True)
```

```python
# Checking dataframe
weather1.info()
```

```python
# Check data types
weather1.dtypes, weather1.index.dtype
```

```python
# so the index is object data-type, which needs to be converted into timeseries
weather1.index = pd.to_datetime(weather1.index)
weather1.index.dtype
```

```python
weather2.boxplot(column=['dew', 'pressure', 'temperature'])
```

```python
sel_cols = ['humidity', 'temperature']
weather3 = weather2.loc[:,sel_cols]
weather3.plot(subplots=True, figsize=(20,12))
```

```python
weather3['2012':'2014'].plot(subplots=True, figsize=(20,12))
```

```python
# drop the humidity data and only focus on temperature from here onwards
weather3.drop(labels='humidity', axis=1, inplace=True)
```

```python
# Divide the data into training and test data.
# Separate all the data till 2014 as training data and rest as test data
weather_train = weather3[:'2014']
weather_test = weather3['2014':]
print("Train Data Elements: ", weather_train.shape[0])
print("Test Data Elements : ", weather_test.shape[0])
```

```python
# check rolling mean and rolling standard deviation
def plot_rolling_mean_std(ts):
    rolling_mean = ts.rolling(12).mean()
    rolling_std = ts.rolling(12).std()
    plt.figure(figsize=(22,10))

    plt.plot(ts, label='Actual Mean')
    plt.plot(rolling_mean, label='Rolling Mean')
    plt.plot(rolling_std, label = 'Rolling Std')
    plt.xlabel("Date")
    plt.ylabel("Mean Temperature")
    plt.title('Rolling Mean & Rolling Standard Deviation')
    plt.legend()
    plt.show()
```

```python
# Augmented Dickey–Fuller test
def perform_dickey_fuller_test(ts):
    result = adfuller(ts, autolag='AIC')
    print('Test statistic: ' , result[0])
    print('Critical Values:' ,result[4])
```

```python
# check stationary: mean, variance(std)and adfuller test
plot_rolling_mean_std(weather_train.temperature)
perform_dickey_fuller_test(weather_train.temperature)
```

```python
# Original Series
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(weather_train.values); 
axes[0, 0].set_title('Original Series')
plot_acf(weather_train.values, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(weather_train.temperature.diff().values); 
axes[1, 0].set_title('1st Order Differencing')
plot_acf(weather_train.diff().dropna().values,ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(weather_train.diff().diff().values); 
axes[2, 0].set_title('2nd Order Differencing')
plot_acf(weather_train.diff().diff().dropna().values,ax=axes[2, 1])

plt.xticks(rotation='vertical')
plt.show()
```

```python
# PACF plot of 1st differenced series
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(weather_train.diff().values); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,5))
plot_pacf(weather_train.diff().dropna().values, ax=axes[1])

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(weather_train.diff().values); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,1.2))
plot_acf(weather_train.diff().dropna().values, ax=axes[1])

plt.show()
```

```python
acf_lag = acf(weather_train.diff().dropna().values, nlags=20)
pacf_lag = pacf(weather_train.diff().dropna().values, nlags=20, method='ols')

plt.figure(figsize=(22,10))

plt.subplot(121)
plt.plot(acf_lag)
plt.axhline(y=0,linestyle='--',color='silver')
plt.axhline(y=-1.96/np.sqrt(len(weather_train.diff().values)),linestyle='--',color='silver')
plt.axhline(y=1.96/np.sqrt(len(weather_train.diff().values)),linestyle='--',color='silver')
plt.title("Autocorrelation Function")

plt.subplot(122)
plt.plot(pacf_lag)
plt.axhline(y=0,linestyle='--',color='silver')
plt.axhline(y=-1.96/np.sqrt(len(weather_train.diff().values)),linestyle='--',color='silver')
plt.axhline(y=1.96/np.sqrt(len(weather_train.diff().values)),linestyle='--',color='silver')
plt.title("Partial Autocorrelation Function")
plt.tight_layout()
```

```python
model = ARIMA(weather_train.values, order=(2,0,2))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
# Plot residual errors
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

```python
# Actual vs Fitted
model_fit.plot_predict(dynamic=False)
plt.show()
```
